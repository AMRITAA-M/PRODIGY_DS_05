```python
import pandas as pd

# Load dataset
file_path = "traffic_accidents.csv"  # Ensure this file is in your working directory
df_uploaded = pd.read_csv(file_path)

# Display basic info to confirm it's loaded
df_uploaded.info()
df_uploaded.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing accidents by time of day
plt.figure(figsize=(8, 5))
sns.countplot(x="Time_of_Day", data=df_uploaded, order=["Morning", "Afternoon", "Evening", "Night"], palette="viridis")
plt.title("Accidents by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the number of accidents per weather condition
weather_counts = df_uploaded["Weather"].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(weather_counts, labels=weather_counts.index, autopct="%1.1f%%", colors=sns.color_palette("coolwarm", len(weather_counts)))
plt.title("Accidents by Weather Condition")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count occurrences of each Road Condition and Severity combination
road_severity_counts = df_uploaded.groupby(["Road_Condition", "Severity"]).size().unstack()

# Plot stacked bar chart
road_severity_counts.plot(kind="bar", stacked=True, figsize=(8, 5), colormap="coolwarm")
plt.title("Accidents by Road Condition and Severity (Stacked)")
plt.xlabel("Road Condition")
plt.ylabel("Number of Accidents")
plt.legend(title="Severity")
plt.xticks(rotation=45)
plt.show()
```
